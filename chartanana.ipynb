```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the CSV file
data = pd.read_csv('dataset/predictions/bia_pred.csv')
```

```python
# Group the data by "Depth"
grouped_data = data.groupby('Palette Size')

# Create a line chart for each depth group
for depth, group in grouped_data:
    plt.plot(group['Depth'], group['Accuracy'], label=f'Palette Size {depth}')


# Add labels and title
plt.xlabel('Depth')
plt.ylabel('Accuracy')
plt.title('Accuracy vs Palette Size')
plt.grid(True)

# Add a legend
plt.legend()

# Display the chart
plt.show()
```
